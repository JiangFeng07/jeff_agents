```python
import os
from langchain_tavily import TavilySearch
from langchain.agents import create_agent
from dotenv import load_dotenv
load_dotenv()

os.environ["LANGSMITH_PROJECT"]="react"
```

```python
search_tool = TavilySearch(max_results=3, include_raw_response=True)

model = "gpt-5-mini"
model = "deepseek:deepseek-chat"
agent = create_agent(
    model=model,
    tools=[search_tool],
    system_prompt="""You are a helpful assistant that can use the provided search tool to answer questions. Use the search tool to find relevant information and provide accurate answers to the user's queries. Always use the search tool when you need to find information to answer the user's question.""",
)
```

```python
for step in agent.stream(
    {
        "messages":[
            {"role": "user", "content": "what is the hometown of the 2025 US Open champion (tennis)?"}        ]
    },
    stream_mode="values"
):
    step["messages"][-1].pretty_print()
```
